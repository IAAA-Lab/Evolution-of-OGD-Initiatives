```python
import json
retweet_list = list()
reply_list = list()
like_list = list()
quote_list = list()
for row in Final_df['NI']:
    row = row.replace("'","\"")
    row = json.loads(row)
    retweet_list.append(row['retweet_count'])
    reply_list.append(row['reply_count'])
    like_list.append(row['like_count'])
    quote_list.append(row['quote_count'])
```

```python
new_df = pd.DataFrame()
new_df['retweet_count'] = retweet_list
new_df['reply_count'] = reply_list
new_df['like_count'] = like_list
new_df['quote_count'] = quote_list
new_df
```

```python
new_df['NI'] = new_df['retweet_count'] + new_df['reply_count'] + new_df['like_count'] + new_df['quote_count']
print(new_df.to_string())
```

```python
df_ti = Final_df.copy()
df_ti['NI'] = new_df['NI']
df_ti
```
